```python
# Zad 1 

print("Prosty kalkulator")

wynik = None
print("Jaką operację chcesz wykonać?")

while True:
    print("Wprowadź odpowiedni znak\nDodawanie - d\nOdejmowanie - o\nMnożenie - m\nDzielenie - z\nZakończ - q\n")
    operacja = input()

    if operacja == "d" or operacja == "o" or operacja == "z" or operacja == "m" and operacja != "q":
        if wynik == None:
            print("Wprowadź pierwszą liczbę: ")
            number1 = int(input())
            print("Wprowadź drugą liczbę: ")
            number2 = int(input())
        else:
            number1 = wynik
            print("Wprowadź liczbę: ")
            number2 = int(input())
    
    if operacja == "d":
        wynik = number1+number2
        print(f"Wynik dodawania liczb: {wynik}")
        
    elif operacja == "o":
        wynik = number1-number2
        print(f"Wynik odejmowania liczb: {wynik}")
        
    elif operacja == "m":
        wynik = number1*number2
        print(f"Wynik mnożenia liczb: {wynik}")
        
    elif operacja == "z":
        wynik = number1/number2
        print(f"Wynik dzielenia liczb: {wynik}")
        
    elif operacja == "q":
        break
        
    else:
        print("Niezrozumiana komenda\n")
```
