W pętli obsługa klientów
Jakie nominały daje klient
Wydawanie reszty (przy możliwie najmniejszej ilości nominałów)

Przejście do kolejnej obsługi klienta

Uniknąć wydawania nominałów, których nie posiadamy


Na 3 - zrobić przy pomocy algorytmu zachłannego 
Na 5 - lepszy algorytm (zastosowanie np. programowania dynamicznego)

Problemy
- problem struktur danych 
- algorytm 

Artmetyka (10 groszy to nie 1/10)

Na koniec pracy generowanie raportu. 

```python
# Symulacja działania kasy sklepowej z algorytmem zachłannym

def wydaj_reszte(reszta, kasa):
    wydane = {}
    nominaly = sorted(kasa.keys(), reverse=True)

    for nom in nominaly:
        ilosc = min(reszta // nom, kasa[nom])
        if ilosc > 0:
            wydane[nom] = ilosc
            reszta -= nom * ilosc

    if reszta == 0:
        return wydane
    else:
        return None


def aktualizuj_kase(kasa, nominaly, dodaj=True):
    for nom, ilosc in nominaly.items():
        kasa[nom] += ilosc if dodaj else -ilosc


def wyswietl_stan_kasy(kasa):
    print("\nAktualny stan kasy:")
    for nom, ilosc in sorted(kasa.items(), reverse=True):
        print(f"Nominał {nom / 100:.2f} zł: {ilosc} szt.")


def symuluj():
    kasa = {
        50000: 2, 20000: 3, 10000: 5, 5000: 10, 2000: 20,
        1000: 50, 500: 50, 200: 100, 100: 100,
        50: 200, 20: 200, 10: 200, 5: 100, 2: 100, 1: 100
    }
    wyswietl_stan_kasy(kasa)

    while True:
        print("\nNowy klient")
        try:
            kwota_do_zaplaty = int(float(input("Podaj kwotę do zapłaty (np. 15.50): ")) * 100)
        except ValueError:
            print("Niepoprawna kwota, spróbuj ponownie.")
            continue

        while True:
            klient_nominaly_input = input("Jakie nominały daje klient? (np. 500zl,200zl): ").replace(' ', '').lower()
            klient_nominaly = []
            try:
                for nom in klient_nominaly_input.split(','):
                    if 'zl' in nom:
                        wartosc = int(float(nom.replace('zl', '')) * 100)
                        klient_nominaly.append(wartosc)
                    else:
                        raise ValueError
                break
            except ValueError:
                print("Wprowadzono niepoprawny nominał, spróbuj ponownie.")

        suma_klienta = sum(klient_nominaly)
        reszta = suma_klienta - kwota_do_zaplaty

        suma_klienta = sum(klient_nominaly)
        resztareszta = suma_klienta - kwota_do_zaplaty

        if reszta < 0:
            print("Klient dał za mało pieniędzy!")
            continue

        if reszta == 0:
            print("Klient zapłacił odliczoną kwotę, nie trzeba wydawać reszty.")
            aktualizuj_kase(kasa, {nom: klient_nominaly.count(nom) for nom in set(klient_nominaly)}, dodaj=True)
        else:
            wydane = wydaj_reszte(reszta, kasa)

            if wydane:
                aktualizuj_kase(kasa, {nom: klient_nominaly.count(nom) for nom in set(klient_nominaly)}, dodaj=True)
                print("Wydaj resztę:")
                for nom, ilosc in wydane.items():
                    print(f"{nom / 100:.2f} zł: {ilosc} szt.")

                aktualizuj_kase(kasa, wydane, dodaj=False)
            else:
                print("Nie można wydać reszty! Transakcja anulowana.")


        wyswietl_stan_kasy(kasa)

        if input("Czy kolejny klient? (tak/nie): ").lower() != 'tak':
            break

    print("\n=== Raport końcowy ===")
    wyswietl_stan_kasy(kasa)


if __name__ == "__main__":
    symuluj()
```

```python
# Symulacja działania kasy sklepowej z algorytmem dynamicznym

def wydaj_reszte(reszta, kasa):
    nominaly = sorted(kasa.keys())
    max_reszta = reszta + 1
    dp = [max_reszta] * (reszta + 1)
    dp[0] = 0
    wybor = [-1] * (reszta + 1)

    for i in range(1, reszta + 1):
        for nom in nominaly:
            if nom <= i and kasa[nom] > 0 and dp[i - nom] + 1 < dp[i]:
                dp[i] = dp[i - nom] + 1
                wybor[i] = nom

    if dp[reszta] == max_reszta:
        return None

    wydane = {}
    i = reszta
    tymczasowa_kasa = kasa.copy()

    while i > 0:
        nom = wybor[i]
        if nom == -1 or tymczasowa_kasa[nom] <= 0:
            return None
        wydane[nom] = wydane.get(nom, 0) + 1
        tymczasowa_kasa[nom] -= 1
        i -= nom

    return wydane


def aktualizuj_kase(kasa, nominaly, dodaj=True):
    for nom, ilosc in nominaly.items():
        kasa[nom] += ilosc if dodaj else -ilosc


def wyswietl_stan_kasy(kasa):
    print("\nAktualny stan kasy:")
    for nom, ilosc in sorted(kasa.items(), reverse=True):
        print(f"Nominał {nom / 100:.2f} zł: {ilosc} szt.")


def symuluj():
    kasa = {50000: 2, 20000: 3, 10000: 5, 5000: 10, 2000: 20, 1000: 50, 500: 50, 200: 100, 100: 100, 50: 200, 20: 200, 10: 200}
    wyswietl_stan_kasy(kasa)

    while True:
        print("\nNowy klient")
        try:
            kwota_do_zaplaty = int(float(input("Podaj kwotę do zapłaty (np. 15.50): ")) * 100)
        except ValueError:
            print("Niepoprawna kwota, spróbuj ponownie.")
            continue

        while True:
            klient_nominaly_input = input("Jakie nominały daje klient? (np. 500zl,200zl): ").replace(' ', '').lower()
            klient_nominaly = []
            try:
                for nom in klient_nominaly_input.split(','):
                    if 'zl' in nom:
                        wartosc = int(float(nom.replace('zl', '')) * 100)
                        klient_nominaly.append(wartosc)
                    else:
                        raise ValueError
                break
            except ValueError:
                print("Wprowadzono niepoprawny nominał, spróbuj ponownie.")

        suma_klienta = sum(klient_nominaly)
        reszta = suma_klienta - kwota_do_zaplaty

        if reszta < 0:
            print("Klient dał za mało pieniędzy!")
            continue

        aktualizuj_kase(kasa, {nom: klient_nominaly.count(nom) for nom in set(klient_nominaly)}, dodaj=True)

        wydane = wydaj_reszte(reszta, kasa)

        if wydane:
            print("Wydaj resztę:")
            for nom, ilosc in wydane.items():
                print(f"{nom / 100:.2f} zł: {ilosc} szt.")

            aktualizuj_kase(kasa, wydane, dodaj=False)
        else:
            print("Nie można wydać reszty!")

        wyswietl_stan_kasy(kasa)

        if input("Czy kolejny klient? (tak/nie): ").lower() != 'tak':
            break

    print("\n=== Raport końcowy ===")
    wyswietl_stan_kasy(kasa)


if __name__ == "__main__":
    symuluj()
```
